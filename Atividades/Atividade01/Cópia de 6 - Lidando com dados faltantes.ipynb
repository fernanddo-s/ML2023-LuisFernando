## Identificando valores faltantes em dados tabulares
Vamos iniciar nosso estudo identificando dados faltantes em tabelas.

Carregando dados de um arquivo .csv

```python
import pandas as pd

df = pd.read_csv('missing.csv')
df
```

Contando quantas ocorrencias de null tem em cada coluna, como as colunas A e B não tem nenhum elas tem 0, em contra partida as colunas C e D tem uma ocorrencia cada uma.

```python
df.isnull().sum()
```

Cria um array numpy com todos os valores do DataFrame, que foi criado anteriormente nas céluas.

```python
# Cria um array numpy com todos os valores do DataFrame

df.values
```

## Eliminando amostras ou *features* com valores ausentes

Remove linhas que contém valores faltantes, nesse caso existem valores faltantes nas linas de indice 1 e 2, nas colunas C e D respectivamente. 

```python
# Remove colunas que contém valores faltantes

df.dropna(axis=0)
```

Removendo colunas que contém valores faltantes, as colunas C e D possuem valores faltantes.

```python
# Removendo colunas que contém valores faltantes

df.dropna(axis=1)
```

"Drop" apenas nas linhas onde todas as colunas são NaN, como nenhuma das linhas possui apenas NaN, todas as colunas ainda aparecem.

```python
# "Drop" apenas nas linhas onde todas as colunas são NaN

df.dropna(how='all')
```

Removendo linhas que possuem menos de 3 valores reais, nesse caso como as linhas 1 e 2 tem um NaN cada uma elas não aparecem no resultado.

```python
# Removendo linhas que possuem menos de 3 valores reais

df.dropna(thresh=4)
```

Remove apenas linhas onde NaN aparece em uma coluna específica. Como a coluna C possui um NaN na linha 1, a linha não aparece no resultado. Ficam aparentes somente as linhas 0 e 2.

```python
# Remove apenas linhas onde NaN aparece em uma coluna específica

df.dropna(subset=['C'])
```
