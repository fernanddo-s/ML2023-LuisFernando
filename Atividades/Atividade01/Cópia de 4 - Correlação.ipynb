# 4 - Correlação
**Direto ao ponto**: correlação é o coeficiente de interdependência entre duas ou mais variáveis.

Em muitos casos, quando analisamos algumas variáveis, essa interdependência é muito clara. Exemplo:
- Massa corporal e altura,
- Consumo de combustível e distância percorrida,
- Número de anúncios e produtos vendidos,
- Tempo de estudo e desempenho na avaliação.

Entretanto, algumas vezes essa correlação pode não ser muito clara.

Correlação procura entender é como uma variável se comporta em um cenário onde outra está variando, visando identificar se existe alguma relação entre a variabilidade de ambas. Embora não implique em causalidade, o coeficiente de correlação pode expressar em números essa relação, ou seja, quantifica a relação entre as variáveis.

0.9 para mais ou para menos indica uma correlação muito forte.

0.7 a 0.9 positivo ou negativo indica uma correlação forte.

0.5 a 0.7 positivo ou negativo indica uma correlação moderada.

0.3 a 0.5 positivo ou negativo indica uma correlação fraca.

## Correlações positivas

Quando  o coeficiente de correlação se aproxima de 1, é possível perceber um aumento no valor de uma variável quando a outra também aumenta, ou seja, há uma relação linear positiva.

```python
import numpy as np

np.random.seed(1)

# 1000 random integers between 0 and 50
x = np.random.randint(0, 50, 1000)

# Positive Correlation with some noise
y = x + np.random.normal(0, 10, 1000)

np.corrcoef(x, y)
```

```python
# import matplotlib
import matplotlib.pyplot as plt

plt.scatter(x, y)
```

## Correlações negativas

Quando o coeficiente se aproxima de -1, é possível dizer que as variáveis são correlacionadas, mas nesse caso quando o valor de uma variável aumenta o da outra diminui sendo assim uma correlação negativa.

```python
# 1000 random integers between 0 and 50
x = np.random.randint(0, 50, 1000)

# Negative Correlation with some noise
y = 100 - x + np.random.normal(0, 5, 1000)

np.corrcoef(x, y)
```

```python
plt.scatter(x, y)
```

## Correlação fraca ou sem correlação

Correlações fracas estão entre 0.3 e 0.5, abaixo de 0.3 não existe correalção.

```python
x = np.random.randint(0, 50, 1000)
y = np.random.randint(0, 50, 1000)

np.corrcoef(x, y)
```

```python
plt.scatter(x, y)
```

## Correlação é diferente de causalidade!
